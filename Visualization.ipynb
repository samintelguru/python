# Python libraries for visualization

1. Matplotlib
- is the mother of visualization in python
- Built from a language called matlab which is a statistical language

2. Seaborn
- It is a library that offers better customization of visualizations that give them a better aesthetic and visual appeal

3. Plotly
- It is a library that introduces an aspect of interaction with visualizations

1. Bar chart
- Uses bars, (columns) to compare discrete categories of data. The bars can either be horizontal or vertical

Best for:
- Comparing values across categories
- Displaying part-to-whole relationships in data

eg. sales  per region, population by age group

2. Line chart
- plots data points along a continuous line, often used to show trend over time

Best for
- Time series analysis (eg. stock prices)
- Showing trends, patterns and changes or intervals

3. Pie charts
- Circular chart divided into slices to show propotions of a whole

Best for:
- Showing percentages or propotional data for a small number of categories (max=5)

4. Scatter plot
- This plots points on the x-y axes to show relationship between two variables

Best for:
- Identifying correlations, clusters, or outliers
- Showing how two variables are related eg. weight vs height

5. Histogram
- Groups together continuous data into bins to show frequency distribution

Best for:
- Visualizing the distribution of a single continuous variable
- Spotting patterns, like skewness and data spread

6. Heatmap
- Uses color to represent values in a grid or matrix format

Best for: Displaying data density, concentration and intensity
- Analyzing patterns across categories and time, often in matrix form


7. Box plot (box and whisker plot)

Shows data distribution through quartiles, highlighting median and outliers

Best for:
- Identifying spread and skewness in a dataset
- Comparing distributions across multiple groups
- Identifying the outliers in a dataset

8. Area chart
- similar a line chart but with the area between the line and the x axis filled in

Best for:
- Showing cumulative trends over time
- Visualizing part to whole relationships over time


9. Bubble chart
- Extension of the scatter plot adding a third dimension with bubble size

```python
import matplotlib.pyplot as plt

x = [5, 10, 15, 20, 25]
y1 = [3, 9, 27, 18, 15]
y2 = [3, 2, 15, 6, 9]

plt.plot(x, y1, label="A")
plt.plot(x, y2, label="B")

plt.xlabel("x axis numbers")
plt.ylabel("y axis numbers")
plt.title("Sample Line Chart")

plt.legend()

plt.show()
```

Annotations - they can be used to add text, arrows or other forms to a plot. They are used to draw attention to key areas or elements in a plot

- The first parameter is the text or the word you want displayed on the chart or visualization or graph
- The second parameter is is xy, which denotes the place you need to highlight in the graph
- The third parameter is xytext, which shows where the word should be placed
- arrowprops facecolor for the color of the arrow and shink is the distance between the arrow point and the line point

```python
x = [1, 2, 3, 4, 5]
y = [3, 9, 27, 18, 15]

plt.plot(x, y, label="Data")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line chart example")

plt.annotate("Highest point", xy=(3, 27), xytext=(3, 9), arrowprops=dict(facecolor="green", shrink=0.2))

plt.annotate("Low", xy=(1, 3), xytext=(1, 10), arrowprops=dict(facecolor="blue", shrink=0.05))

plt.show()
```

Line styles - specifies the appearance of a line plot. It is used to modify the line's appearance such as its thickness, color and dashes

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]
y3 = [2, 5, 8, 12, 15]
y4 = [5, 5, 5, 7, 8]
y5 = [10, 5, 3, 10, 16]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2", linestyle="--")
plt.plot(x, y3, label="Line 3", linestyle=":")
plt.plot(x, y4, label="Line 4", linestyle="dotted")
plt.plot(x, y5, label="Line 5", linestyle="solid")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Customized line graph")

plt.legend()

plt.show()
```

Adding a marker in matplotlib

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2", marker="x")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line plot with marker")

plt.legend()

plt.show()
```

Setting axes - in matplotlib, by setting values on the xlim and ylim functions, the graph will only focus on these points (will not show the full graph)

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [3, 4, 5, 6, 7]

plt.plot(x, y1, label="Line 1", marker="v")
plt.plot(x, y2, label="Line 2", marker="8")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line plot with x and y limitations")

plt.legend()

plt.xlim(1.8, 4.5)
plt.ylim(2, 8)

plt.show()
```

Setting grid background - the grid is a collection of lines drawn behind the data points in a plot. You can change the linestyle to solid, dashed, dotted, dashdot and "". Then set alpha for the transparency value

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 2, 6, 12]
y2 = [3, 12, 15, 6, 9]

plt.plot(x, y1, label='A', color='#506ef0')
plt.plot(x, y2, label='B', color='#bf5740', marker='o')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Grid lines')

plt.legend()

plt.grid(True, linestyle='-.', alpha=0.9, linewidth=0.5)
#plt.grid(True, linestyle='dashed', alpha=1, linewidth=1)

plt.show()
```

Setting the background color

The set_facecolor() can be used to change the background color

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 2, 6, 12]
y2 = [3, 12, 15, 6, 9]

plt.plot(x, y1, label='A',color='#3559ca')
plt.plot(x, y2, label='B', color='#c23d8b', marker='o')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Background color')

plt.legend()

# setting the background color to white
plt.gca().set_facecolor('#ffffff')

plt.show()
```

Subplots

Subplots are used to create numerous plots within a single figure. This lets you structure and compare multiple datasets, visualizations or features in your data

Saving a figure

The savefig() allows you to save your plots as picture files. This allows you to save your visualizations for use in presentations or report

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 2, 6, 12]
y2 = [3, 12, 15, 6, 9]
y3 = [4, 5, 7, 9, 10]
y4 = [3, 4, 4 ,6, 9]

fig, plotting = plt.subplots(nrows=2, ncols=3, figsize=(15, 6))

plotting[0, 0].plot(x, y1, label="Line1")
plotting[0, 0].set_title("Line 1")

plotting[0, 1].plot(x, y2, label="Line2")
plotting[0, 1].set_title("Line 2")

plotting[0, 2].plot(x, y2, label="Line3")
plotting[0, 2].set_title("Line 3")

plotting[1, 0].plot(x, y3, label="Line4")
plotting[1, 0].set_title("Line 4")

plotting[1, 1].plot(x, y4, label="Line5")
plotting[1, 1].set_title("Line 5")

plotting[1, 2].plot(x, y4, label="Line6")
plotting[1, 2].set_title("Line 6")

plt.savefig("subplots.png")

plt.show()


```

```python
import random

x = []
y = []
sizes = []

for i in range(15):
  x_num = random.randint(1, 15)
  y_num = random.randint(1, 15)
  s_num = random.randint(1, 100)

  x.append(x_num)
  y.append(y_num)
  sizes.append(s_num)

plt.figure(figsize=(8, 5))

plt.scatter(x, y, s=sizes, c="#2b073d", alpha=0.5)

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sample scatter plot")

plt.show()
```

Relational plots

- are mostly plotted for bivariate analysis.By plotting these, wer try to understand the statistical relationship between two or more features.



Scatterplot

```python
# figure level plot

sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter")
plt.title("Scatter plot showing the Total bill vs Tip")
plt.show()
```

```python
# axes level plot

sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter plot showing the Total bill vs Tip")
plt.xlabel("Total bill")
plt.ylabel("Tip")
plt.show()
```

Note : Graphs plotted using the figure level function have a squarish outlook while axes level function often produces a graph that is rectangualar in shape after the default auto adjustments.

KDE Plot - Kernel Distribution Estimation - is a plot used to estimate the probability density function of a continuous variable

It probives a smoothed representation of the distribution of data

Categorical plots

allow us to visualize the distribution of data across categorical variables. These plots are particularly useful when we want to explore the relationship between one categorical variable and one or more numerical variables

Seaborn has 6 types of categorical plots
1. Bar plot
2. Count plot
3. Box plot
4. Violin plot
5. Swarm plot
6. Point plot

Bar plot

```python
g = sns.barplot(data=tips, x="day", y="total_bill", estimator=np.mean)
g.set_title("Daywise total bill")
```

```python
sns.catplot(kind="bar", data=tips, x="day", y="total_bill", estimator=np.median, errorbar=None)
```
