```python
goals_per_game = df.groupby("Player").agg(
    total_goals = ("Player", "count"),
    total_games = ("Date","nunique")

)
goals_per_game["goals_per_game"] = goals_per_game["total_goals"]/goals_per_game["total_games"]
goals_per_game
```

What is the scoring rate for each of the two players per competition?

What is the scoring rate for each of thr two players per season?


```python
goals_per_competitions = df.groupby(["Player", "Competition"]).agg(
    total_goals = ("Player", "count"),
    total_competitions = ("Competition","nunique")

)
goals_per_competitions["goals_per_competition"] = goals_per_competitions["total_goals"]/goals_per_competitions["total_competitions"]
goals_per_competitions

```

```python
goals_per_season = df.groupby(["Player", "Season"]) .agg(
    total_goals = ("Player", "count"),
    total_seasons = ("Season","nunique")

)
goals_per_season["goals_per_season"] = goals_per_season["total_goals"]/goals_per_season["total_seasons"]
goals_per_season
```

- What is the scoring rate for each of the players per day of the week?
- What is the scoring rate for each of the players per month?

```python
goals_per_day_of_week = df.groupby(["day_name", "Player"]) .agg(
    total_goals = ("Player", "count"),
    total_games= ("Date","nunique")

)
goals_per_day_of_week["goals_per_day_of_week"] = goals_per_day_of_week["total_goals"]/goals_per_day_of_week["total_games"]
goals_per_day_of_week.sort_values("goals_per_day_of_week", ascending=False)
```

```python
goals_per_month = df.groupby(["month", "Player"]) .agg(
    total_goals = ("Player", "count"),
    total_games= ("Date","nunique")

)
goals_per_month["goals_per_month"] = goals_per_month["total_goals"]/goals_per_month["total_games"]
goals_per_month.sort_values(by = "goals_per_month", ascending=False)
```
